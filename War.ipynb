# Game 2: WAR

Game Rules:
1. Deck is shuffled and divided among 2 players
2. Each player draws one card at the top of stack and compares it with the opponent
3. Player with higher card value wins that round and takes both the cards
4. If both players draw cards of same value, then they are at WAR
5. If at WAR, players will grab another card and continue the WAR
6. This goes on until one player is out of cards
7. Player who is out of cards loses the game

```python
import random
```

```python
# initialising cards
suits = ('Hearts','Diamonds','Clubs','Spades')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':11,'Queen':12,'King':13,'Ace':14}
```

```python
# create a card class
# assign each card a value
# return the suit with it's rank
class Card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
# create a deck class
class Deck:
    def __init__(self):
        # happens only upon creation of new deck
        self.all_cards = []
        for s in suits:
            for r in ranks:
                self.all_cards.append(Card(s,r))
    
    def shuffle(self):
        # shuffles the deck
        random.shuffle(self.all_cards)
    
    def deal_one(self):
        # returns one card from the list
        return self.all_cards.pop()
```

```python
# create a player class
class Player:
    def __init__(self,name):
        self.name = name
        # initially new player has no cards
        self.all_cards = []
        
    def remove_one(self):
        # we remove a card form top of stack
        return self.all_cards.pop(0)
    
    def add_cards(self,newCards):
        if type(newCards) == type([]):
            self.all_cards.extend(newCards)
        else:
            self.all_cards.append(newCards)
            
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards'
```

```python
# Game Setup
new_deck = Deck()
new_deck.shuffle()

p1 = Player('One')
p2 = Player('Two')

for x in range(26):
    p1.add_cards(new_deck.deal_one())
    p2.add_cards(new_deck.deal_one())
    
import pdb
game_on = True
```

```python
# Play the game
round_num = 0
while game_on:
    
    round_num += 1
    print(f'Round {round_num}')
    
    # check if player is out of cards
    if(len(p1.all_cards)) == 0:
        print('Player One out of cards, Game over!')
        print('Player Two wins!')
        game_on = False
        break
    if(len(p2.all_cards)) == 0:
        print('Player Two out of cards, Game over!')
        print('Player One wins!')
        game_on = False
        break
        
    # otherwise
    
    #start new round, reset current cards
    p1_cards = []
    p1_cards.append(p1.remove_one())
    
    p2_cards = []
    p2_cards.append(p2.remove_one())
    
    at_war = True
    
    while at_war:
        
        if p1_cards[-1].value > p2_cards[-1].value:
            # p1 gets all the cards
            p1.add_cards(p1_cards)
            p1.add_cards(p2_cards)
            
            # no longer at war
            at_war = False
            
        elif p1_cards[-1].value < p2_cards[-1].value:
            # p1 gets all the cards
            p2.add_cards(p1_cards)
            p2.add_cards(p2_cards)
            
            # no longer at war
            at_war = False
            
        else:
            # when cards are equal
            # grab another card and continue the war
            print('WAR')
            
            # check is player is out of cards
            if len(p1.all_cards) < 5:
                print('Player One is out of cards. Game over at WAR')
                print('Player Two wins the WAR!')
                game_on = False
                break
            
            elif len(p2.all_cards) < 5:
                print('Player Two is out of cards. Game over at WAR')
                print('Player One wins the WAR!')
                game_on = False
                break
            
            # otherwise we're still at war
            else:
                for n in range(5):
                    p1_cards.append(p1.remove_one())
                    p2_cards.append(p2.remove_one())
```
